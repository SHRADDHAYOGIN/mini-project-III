```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_df = pd.read_csv('../data/twm_customer.csv', sep=';')
customer_df.info()
```

```python
customer_df = customer_df[['cust_id', 'income', 'years_with_bank']]
customer_df.head()
```

```python
accounts_df = pd.read_csv('../data/twm_accounts.csv', sep=';') 
accounts_df.head()
```

```python
tnx_df = pd.read_csv('../data/twm_transactions.csv', sep=';') 
tnx_df.head()
```

### Savings
#### Savings Account

```python
savings_accounts_df = pd.read_csv('../data/twm_savings_acct.csv', sep=';') 
# accounts_df.head()

customer_accounts_df = customer_df.merge(savings_accounts_df, 
#                                          how='left',
                                         left_on='cust_id', 
                                         right_on='cust_id')
customer_accounts_df.sort_values('cust_id')
```

```python
customer_accounts_df = customer_accounts_df[['cust_id', 'income', 'years_with_bank',
                                            'starting_balance', 'ending_balance',
                                            'account_active']]
customer_accounts_df.info()
```

```python
customer_accounts_df['hasSavings'] = customer_accounts_df['starting_balance'].\
                                    apply(lambda x: 'Yes' if x else 'No')
customer_accounts_df
```

```python
# category for income saved
customer_accounts_df['incomeSavedOverTime'] = \
    (customer_accounts_df['income'] * customer_accounts_df['years_with_bank']) \
    /customer_accounts_df['ending_balance']
customer_accounts_df
```

#### Savings transactions

```python
savings_tnx_df = pd.read_csv('../data/twm_savings_tran.csv', sep=';') 
# accounts_df.head()

customer_accounts_tnx_df = customer_accounts_df.merge(savings_tnx_df, 
#                                          how='left',
                                         left_on='cust_id', 
                                         right_on='cust_id')
customer_accounts_tnx_df.sort_values('cust_id')
```

```python
# category number of transactions.
num_of_tnx = customer_accounts_tnx_df.groupby('cust_id')['tran_date'].count()
num_of_tnx
```

```python
customer_svngs_df = customer_accounts_tnx_df.groupby('cust_id').agg({
    'hasSavings': pd.Series.mode,
    'incomeSavedOverTime': pd.Series.mode})
customer_svngs_df
```

```python
customer_svngs_df = customer_svngs_df.merge(num_of_tnx, left_index=True, right_index=True)
customer_svngs_df.rename(columns = {'tran_date': 'numSavingsTnx'}, inplace=True)
```

```python
customer_svngs_df
```

## Checking

### Checking Account

```python
checking_accounts_df = pd.read_csv('../data/twm_checking_acct.csv', sep=';') 
# accounts_df.head()


customer_accounts_df = customer_df.merge(checking_accounts_df, 
#                                          how='left',
                                         left_on='cust_id', 
                                         right_on='cust_id')
customer_accounts_df.sort_values('cust_id')
```

```python
customer_accounts_df = customer_accounts_df[['cust_id', 'income', 'years_with_bank',
                                            'starting_balance', 'ending_balance',
                                            'account_active']]
customer_accounts_df
```

```python
customer_accounts_df['hasChecking'] = customer_accounts_df['starting_balance'].\
                                    apply(lambda x: 'Yes' if x else 'No')
customer_accounts_df
```

### Checking transactions

```python
checking_tnx_df = pd.read_csv('../data/twm_checking_tran.csv', sep=';') 
# accounts_df.head()

customer_accounts_tnx_df = customer_accounts_df.merge(checking_tnx_df, 
#                                          how='left',
                                         left_on='cust_id', 
                                         right_on='cust_id')
customer_accounts_tnx_df.sort_values('cust_id')
```

```python
# category number of transactions.
num_of_tnx = customer_accounts_tnx_df.groupby('cust_id')['tran_date'].count()
num_of_tnx
```

```python
# category average transactions
average_tnx = customer_accounts_tnx_df.groupby('cust_id')['tran_amt'].mean()
average_tnx
```

```python
customer_chkng_df = customer_accounts_tnx_df.groupby('cust_id').agg({
    'hasChecking': pd.Series.mode})
customer_chkng_df
```

```python
customer_chkng_df = customer_chkng_df.merge(num_of_tnx, left_index=True, right_index=True)
customer_chkng_df = customer_chkng_df.merge(average_tnx, left_index=True, right_index=True)
customer_chkng_df
```

```python
customer_chkng_df.rename(columns = {'tran_date': 'numCheckingTnx'}, inplace=True)
customer_chkng_df.rename(columns = {'tran_amt': 'averageCheckingTnx'}, inplace=True)
customer_chkng_df
```

### Credit
#### Credit account

```python
credit_accounts_df = pd.read_csv('../data/twm_credit_acct.csv', sep=';') 
# accounts_df.head()


customer_accounts_df = customer_df.merge(credit_accounts_df, 
#                                          how='left',
                                         left_on='cust_id', 
                                         right_on='cust_id')
customer_accounts_df.sort_values('cust_id')
```

```python
customer_accounts_df = customer_accounts_df[['cust_id', 'income', 'years_with_bank',
                                             'credit_limit', 'credit_rating', 
                                            'starting_balance', 'ending_balance',
                                            'account_active']]
customer_accounts_df
```

```python
customer_accounts_df['hasCredit'] = customer_accounts_df['starting_balance'].\
                                    apply(lambda x: 'Yes' if x else 'No')
customer_accounts_df
```

#### Credit transactions

```python
credit_tnx_df = pd.read_csv('../data/twm_credit_tran.csv', sep=';') 
# accounts_df.head()

customer_accounts_tnx_df = customer_accounts_df.merge(credit_tnx_df, 
#                                          how='left',
                                         left_on='cust_id', 
                                         right_on='cust_id')
customer_accounts_tnx_df.sort_values('cust_id')
```

```python
# category number of transactions.
num_of_tnx = customer_accounts_tnx_df.groupby('cust_id')['tran_date'].count()
num_of_tnx
```

```python
# average transactions
for idx in customer_accounts_tnx_df.index:
    td = customer_accounts_tnx_df.loc[idx, 'tran_date']
    customer_accounts_tnx_df.loc[idx, 'tran_month'] = pd.to_datetime(td).month
    
customer_accounts_tnx_df['tran_month'] = customer_accounts_tnx_df['tran_month'].astype(int)
```

```python
tnx_per_month = customer_accounts_tnx_df.groupby(['cust_id', 'tran_month']).mean('tran_amt')
```

```python
tnx_per_month
```

```python
tnx_average = tnx_per_month.groupby(['cust_id']).sum('tran_amt')/tnx_per_month.groupby(['cust_id']).count()
```

```python
tnx_average
```

```python
tnx_average = tnx_average['tran_amt']
tnx_average
```

```python
customer_credit_df = customer_accounts_tnx_df.groupby('cust_id').agg({
    'hasCredit': pd.Series.mode,
    'credit_limit': pd.Series.mode})
customer_credit_df
```

```python
customer_credit_df = customer_credit_df.merge(num_of_tnx, left_index=True, right_index=True)
customer_credit_df = customer_credit_df.merge(tnx_average, left_index=True, right_index=True)
customer_credit_df
```

```python
customer_credit_df.rename(columns = {'tran_date': 'numCreditTnx'}, inplace=True)
customer_credit_df.rename(columns = {'tran_amt': 'averageCreditTnx'}, inplace=True)
customer_credit_df
```

```python
customer_credit_df['creditSpendToLimitRatio'] = \
    customer_credit_df['averageCreditTnx']/customer_credit_df['credit_limit']

customer_credit_df.drop(['averageCreditTnx' , 'credit_limit'], axis=1, inplace=True)
customer_credit_df
```
