```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/customer_final_df')
df
```

```python
df.drop(columns='cust_id', inplace=True)
df.info()
```

```python
df_numerical = ['numCreditTnx', 
                'creditSpendToLimitRatio', 
               'incomeSavedOverTime',
               'averageCheckingTnx']
for col in df_numerical:
    
    plt.hist(df[col], 
#              bins,
             label=col)
    plt.legend(loc='best')
    plt.show()
    
```

```python
# box plot
for col in df_numerical:
    
    plt.boxplot(df[col])
    plt.xlabel(col)
    plt.show()
```

```python
for col in ['hasCredit', 'hasSavings', 'hasChecking']:
    df[col] = df[col].apply(lambda x: 1 if x == 'Yes' else 0)
df
```

```python
df.to_csv('../data/model_final_df', index=False)

pd.read_csv('../data/model_final_df')
```
